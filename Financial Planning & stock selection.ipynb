```python
goal_to_achieve = int(input('Option - the amount you need to achieve for the goal'))
how_many_years = int(input(('How much time will you require your goal amount')))
investment_mode = str(input(('SIP or Lumpsum')))

```

```python
def goal(goal_to_achieve, how_many_years, risk_appettie, investment_mode):
    
    if investment_mode == "Lumpsum":
        if risk_appettie == 'Highly Aggressive':
            ROI = 0.17
            FV = goal_to_achieve
            PV = round(FV/(1+ROI)**how_many_years,0)
            return print(' You need to pay Lumpsum payment of', PV,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        elif risk_appettie == 'Moderately Aggressive':
            ROI = 0.15
            FV = goal_to_achieve
            PV = round(FV/(1+ROI)**how_many_years,0)
            return print(' You need to pay Lumpsum payment of', PV,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        elif risk_appettie == 'Conservative':
            ROI = 0.13
            FV = goal_to_achieve
            PV = round(FV/(1+ROI)**how_many_years,0)
            return print(' You need to pay Lumpsum payment of', PV,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        elif risk_appettie == 'Most Conservative':
            ROI = 0.10
            FV = goal_to_achieve
            PV = round(FV/(1+ROI)**how_many_years,0)
            return print(' You need to pay Lumpsum payment of', PV,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        
    elif investment_mode == 'SIP':
        if risk_appettie == 'Highly Aggressive':
            ROI = 0.17
            I = ROI/12
            Months = how_many_years * 12
            FV = goal_to_achieve
            SIP_amount = ((((1+I)**Months - 1)/ I) * (1+ I))/( FV )
            SIP_amount = round(1 / SIP_amount, 0)
            return print(' You need to pay Monthly payment of', SIP_amount,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        
        elif risk_appettie == 'Moderately Aggressive':
            ROI = 0.15
            I = ROI/12
            Months = how_many_years * 12
            FV = goal_to_achieve
            SIP_amount = (((1+I)**Months - 1)/ I) * (1+ I)/ FV
            SIP_amount = round(1 / SIP_amount, 0)
            return print(' You need to pay Monthly payment of', SIP_amount,' to achieve your goal', FV, 'within', how_many_years ,'Years')
              
        elif risk_appettie == 'Conservative':
            ROI = 0.13
            I = ROI/12
            Months = how_many_years * 12
            FV = goal_to_achieve
            SIP_amount = (((1+I)**Months - 1)/ I) * (1+ I)/ FV
            SIP_amount = round(1 / SIP_amount, 0)
            return print(' You need to pay Monthly payment of', SIP_amount,' to achieve your goal', FV, 'within', how_many_years ,'Years')
    
        elif risk_appettie == 'Most Conservative':
            ROI = 0.10
            I = ROI/12
            Months = how_many_years * 12
            FV = goal_to_achieve
            SIP_amount = (((1+I)**Months - 1)/ I) * (1+ I)/ FV
            SIP_amount = round(1 / SIP_amount, 0)
            return print(' You need to pay Monthly payment of', SIP_amount,' to achieve your goal', FV, 'within', how_many_years ,'Years')
        
```

```python
goal(goal_to_achieve, how_many_years,risk_appettie,investment_mode)
```
